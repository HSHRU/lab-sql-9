```python
#1. How many distinct (different) actors' last names are there?
data1 = pd.read_sql_query('SELECT COUNT(DISTINCT(last_name)) AS different_actor_names FROM sakila.actor', engine)
data1.head()
```

```python
data2 = '''SELECT *
,DATE_FORMAT(rental_date, '%%M') AS 'month'
,DATE_FORMAT(rental_date, '%%W') AS 'weekday'
,CASE
WHEN DATE_FORMAT(rental_date, '%%W') > 5 THEN 'weekend'
ELSE 'workday'
END AS day_type
FROM rental
ORDER BY rental_date;'''
data = pd.read_sql_query(data2, engine)
data.head()
```

```python
#3 Get all films with ARMAGEDDON in the title.
data3 = pd.read_sql_query('SELECT * FROM film WHERE title LIKE "%%ARMAGEDDON%%"', engine)
data3
```

```python
#4 Get 10 the longest films.
data4 = pd.read_sql_query('SELECT title, length FROM film ORDER BY length DESC LIMIT 10', engine)
data4
```

```python

data5 = pd.read_sql_query('select count(*) as "Behind the scenes" from sakila.film where special_features like "%%Behind the Scenes%%" ', engine)
data5
```

```python
#6. Which kind of movies (rating) have a mean duration of more than two hours?
data6 = 'SELECT rating, ROUND(AVG(length),2) as "AVG length" FROM film GROUP BY rating HAVING AVG(length)>120'
hrs_2 = pd.read_sql_query(data6, engine)
hrs_2.head()
```
